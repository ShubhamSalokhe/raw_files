```python
plt.figure(figsize = (5,5))
plt.plot(np.random.randn(30),'mo-',label='FirstPlot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple line plot')
plt.legend()
```

```python
plt.figure()
x = np.arange(10)
y = x**2
z = x**3
plt.plot(x,y,'mo-',label = 'Square',color = 'blue')
plt.plot(x,z,'mo-',label = 'cube')
```

```python
# bar graphs
plt.figure()
x = np.arange(10)
y = x *2 
plt.bar (x,y, width = 0.8)
plt.xticks(x)
plt.yticks(y)
```

```python
plt.figure()
x = np.arange(10)
y = x* 2
z = x*4
plt.bar(x,y, width = 0.4 , color = 'm')
plt.bar(x,z, width = 0.4, color = 'c' ,bottom = x) # stagged bar graph
```

```python
plt.figure()
x = np.arange(10)
y = x* 2
z = x* 4
plt.bar(x-0.2,y, width = 0.4 , color = 'm')
plt.bar(x+0.3,z, width = 0.4 , color = 'b')
```

```python
#histogram shows frequency 
plt.figure()
plt.hist([1,2,3,4,5,4,2,5,6,7,8,9,5,6,7,0,5,6,7])
plt.xticks(np.arange(10))
plt.yticks(np.arange(5))
```
