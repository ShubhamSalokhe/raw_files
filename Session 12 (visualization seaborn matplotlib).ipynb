```python
col1 = np.linspace(0,10,100)
col2 = np.sin(col1)
df = pd.DataFrame({'C1':col1,'C2':col2})
df.head()
```

```python
plt.figure(figsize = (8,6))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = df.C1, y =df.C2)
```

```python
# RC dictionary
plt.figure(figsize = (12,8))
sns.set(rc = {'axes.facecolor':'cyan','axes.grid':True, 'xtick.labelsize':16, 'ytick.labelsize':16})
sns.lineplot(df.C1,df.C2, data= df, linewidth = 5)
sns.lineplot(df.C1,df.C2+1, data= df, linewidth = 5)
```

```python
# RC dictionary
plt.figure(figsize = (12,8))
sns.set(rc = {'axes.facecolor':'#283747','axes.grid':False, 'xtick.labelsize':16, 'ytick.labelsize':16})
sns.lineplot(df.C1,df.C2, data= df, linewidth = 5, color = '#ff9900')
sns.lineplot(df.C1,df.C2+1, data= df, linewidth = 5, color = '#34eb37')
plt.title('cool stuff of seaborn')
```

```python
auto = pd.read_csv('Automobile_data.csv')
auto.head()
```

```python
plt.figure(figsize = (12,8))
sns.lineplot(x = 'wheel-base', y = 'engine-size', data= auto)
```

```python
plt.figure(figsize = (12,8))
sns.lineplot(x = 'wheel-base', y = 'engine-size', data= auto, linewidth = 5, color = 'black')
plt.title('wheel base vs engine size')
```

```python
plt.figure(figsize = (20,7))
sns.lineplot(data = auto['wheel-base'],linewidth = 3, label = 'Wheel Base')
sns.lineplot(data = auto['engine-size'],linewidth = 3, label = 'Engine Size')
```

```python
plt.figure(figsize = (14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = 'wheel-base', y = 'engine-size',hue ='fuel-type', marker = True, data = auto,linewidth  =3)
# hue is based on categories based on color
```

```python
plt.figure(figsize = (14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = 'wheel-base', y = 'engine-size',hue ='body-style',style ='fuel-type', marker = True, data = auto,linewidth  =3)
```

```python
plt.figure(figsize = (14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = 'wheel-base', y = 'engine-size',style ='fuel-type',dashes = False , palette = 'CMRmap',markers = ['o','>'], data = auto,linewidth  =3)
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = 'engine-size', y = 'wheel-base', data= auto , color ='g',marker = '>')
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = 'engine-size', y = 'wheel-base', data= auto ,hue = 'body-style' ,style = 'engine-type',color ='g',marker = '>')
```

```python
# relplot relation between two thing 
plt.figure(figsize = (10,8))
sns.relplot(x = 'engine-size', y = 'wheel-base', data= auto ,kind = 'scatter',hue = 'body-style' ,height = 8,aspect = 1 )
plt.show()
```

```python
plt.figure(figsize=(14,16))
sns.set(rc={'xtick.labelsize':12,'ytick.labelsize':12,'axes.labelsize':12})
sns.relplot(x="wheel-base" , y="engine-size" , hue="drive-wheels" ,kind='line',data=auto, col = 'body-style',col_wrap = 3)
plt.show()
```

```python
plt.figure(figsize=(14,16))
sns.set(rc={'xtick.labelsize':12,'ytick.labelsize':12,'axes.labelsize':12})
sns.relplot(x="wheel-base" , y="engine-size" , hue="drive-wheels" ,kind='line',data=auto, col = 'body-style',col_wrap = 3)
plt.show()
```

```python
plt.figure(figsize=(14,16))
sns.relplot(x="wheel-base" , y="engine-size" , hue="drive-wheels" ,kind='line',data=auto,col = 'num-of-doors',row = 'body-style')
# empty blocks means no cars 
```
